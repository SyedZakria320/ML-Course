```python
import math as M
import numpy as NP
from scipy.spatial import distance_matrix
```

```python
def Euclidean_Distance(A):                  # Function Definition for Distance Calculation
    n = A.shape [0]
    E = NP.zeros( (n, n) )
    for x in range(n):
        for y in range(n):
            E[x, y] = M.sqrt( (abs(A[x, 0]) - abs(A[y, 0]))**2 + (abs(A[x, 1]) - abs(A[y, 1]))**2 )
            E[y, x] = M.sqrt( (abs(A[x, 0]) - abs(A[y, 0]))**2 + (abs(A[x, 1]) - abs(A[y, 1]))**2 )
    return E
```

```python
X = NP.random.normal(size=(200, 2))             # (Positive Integer, 2)
print (X)

MATRIX_loop = Euclidean_Distance(X)            # Symmetric Matrix of Euclidean Distances from loops
print (MATRIX_loop)

MATRIX_vector = distance_matrix(X, X)         # Symmetric Matrix of Euclidean Distances from vectorization
print (MATRIX_vector)

%timeit Euclidean_Distance(X)                  # Time Complexity Analysis of Methods 
%timeit distance_matrix(X, X)
```
